```python
import pandas as pd

df = pd.read_csv('../data/trust_chain_dataset.csv')

print(df.shape)
print(df.head())
print(df.columns)
print(df.dtypes)
```

```python
print(df.describe())
print(df.info())
```

```python
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(10, 3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#  Select only numeric columns
numeric_df = df.select_dtypes(include='number')

#  Compute correlation matrix on numeric data
corr = numeric_df.corr()

#  Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
#  Save cleaned DataFrame to correct folder
df.to_csv("../data/df_cleaned.csv", index=False)

print(" Cleaned dataset saved to: ../data/df_cleaned.csv")
```
